# 9. Exploratory Data Analysis (EDA)

#### 1) Revenue by Product Line

```python
revenue_product = (
    df.groupby("Product_Line")["Revenue"]
    .sum()
    .sort_values(ascending=False)
)

print(revenue_product)
```

```python
#Chart
max_val = revenue_product.max()
min_val = revenue_product.min()

colors = []

for value in revenue_product:
    if value == max_val:
        colors.append("navy")
    elif value == min_val:
        colors.append("red")
    else:
        colors.append("lightblue")

plt.figure(figsize=(6,3))

ax = sns.barplot(
    x=revenue_product.index,
    y=revenue_product.values,
    hue=revenue_product.index,
    palette=colors,
    legend=False
)
for container in ax.containers:
    labels = [f'₹{v:,.0f}' for v in container.datavalues]
    ax.bar_label(container, labels=labels)

ax.yaxis.set_visible(False)

sns.despine(left=True)

plt.title(
    "Revenue by Product Line",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("")
plt.ylabel("")


plt.tight_layout()
plt.show()

```

#### Insight
 Training generated the highest revenue at ₹195,552.89, indicating the strongest customer demand among all product categories. Running generated the lowest revenue at ₹72,075.95, suggesting relatively lower market demand compared to other product lines.

#### 6) Monthly Revenue Trend

```python
#Prepare data
df_time = df.dropna(subset=["Order_Date"])

month_order = [
    "January","February","March","April","May","June",
    "July","August","September","October","November","December"
]

monthly_revenue = (
    df_time.groupby("Month")["Revenue"]
    .sum()
    .reindex(month_order)
)
print(monthly_revenue)
```

```python


#chart
plt.figure(figsize=(8,4))

ax = sns.lineplot(
    x=monthly_revenue.index,
    y=monthly_revenue.values,
    marker="o",
    linewidth=3
)

for x, y in zip(monthly_revenue.index, monthly_revenue.values):
    plt.text(
        x,
        y,
        f'₹{y:,.0f}',
        ha='center',
        va='bottom'
    )

sns.despine()

plt.title(
    "Monthly Revenue Trend",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("")
plt.ylabel("")

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

```

#### 
Insight:
This trend highlights seasonal demand patterns and peak sales periods.

Revenue fluctuated throughout the year, with December (₹73,355.11) generating the highest sales, likely due to holiday and festive season demand. 
February (₹25,250.83) recorded the lowest revenue, indicating a relatively slow sales period.

#### 9) Top Loss-Making Products

```python
# Top 10 loss-making products
top_loss_products = (
    df[df["Profit"] < 0]
    .groupby("Product_Name")["Profit"]
    .sum()
    .sort_values()
    .head(10)
)
    
print(top_loss_products)
```

```python

# Top 10 loss-making products
#chart
plt.figure(figsize=(8,4))

ax = sns.barplot(
    x=top_loss_products.values,
    y=top_loss_products.index,
    color="crimson"
)

# Add loss values at the end of each bar
for i, value in enumerate(top_loss_products.values):
    ax.text(
        value,
        i,
        f" ₹{abs(value):,.0f}",
        va="center",
        ha="left"
    )

plt.title(
    "Top 10 Loss-Making Products",
    fontsize=16,
    fontweight="bold"
)

plt.xlabel("Total Loss (₹)")
plt.ylabel("Product Name")

plt.tight_layout()
plt.show()
```

#### Insight:
 Although these products generated overall profits, they also recorded significant losses from specific transactions. 
Free RN (-₹21,277.95) and Air Jordan (-₹20,834.52) experienced the highest transaction-level losses, suggesting potential issues related to discounting,
returns, or pricing strategies.
